```python
%%writefile app.py
from keras.models import load_model
from tensorflow import keras
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.mobilenet_v2 import preprocess_input
from keras.applications.mobilenet_v2 import decode_predictions
from keras.applications import mobilenet_v2

import streamlit as st 
from PIL import Image
import imageio

def Predict(image1): 
    model = keras.models.load_model('/content/model.hdf5')
    image = imageio.imread(image1, target_size=(224, 224), pilmode="RGB")
    image = img_to_array(image)
    image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
    image = preprocess_input(image)
    base = model.predict(image)
    label = decode_predictions(base)
    label = label[0][0]
    return label

st.title("Image classification")

from tempfile import NamedTemporaryFile
st.set_option('deprecation.showfileUploaderEncoding', False)

temp_file = NamedTemporaryFile(delete=False)

uploaded_file = st.file_uploader("Upload image: ", type="jpg")
if uploaded_file is not None:
    
    temp_file.write(uploaded_file.getvalue())
    image = Image.open(uploaded_file)

    st.image(image, caption='Uploaded Image.', use_column_width=True)
    st.write("")
    if st.button('PREDICT'):
        label = Predict(uploaded_file)
        st.write('%s (%.2f%%)' % (label[1], label[2]*100))
```
